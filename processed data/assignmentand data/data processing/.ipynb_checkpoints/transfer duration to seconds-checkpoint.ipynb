```python
dt.Duration.tolist()[0]
```

```python
def duration_chinese_to_seconds(x):
    if not x is np.nan:
        print(x)
        seconds = 0
        if 'hour' in x:
            hr = int(x[:x.index('hour')])
            #print(hr)
            seconds += hr*60*60
            x = x[x.index('hour')+5:]
            #print(x)
        if "minutes" in x:
            mins = int(x[:x.index("minutes")])
            #print(mins)
            seconds += mins*60
            x = x[x.index("minutes")+7:]
            #print(x)
        if "seconds" in x:
            seconds += int(x[:x.index("seconds")])
        return seconds
```

```python
for i in dt.Duration.tolist():
    print(duration_chinese_to_seconds(i))
```

```python
dt['seconds'] = dt['Duration'].apply(duration_chinese_to_seconds)
```

```python
dt.drop(columns=['Duration'])

```
